```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
file_path = 'C:/Users/Melka/Downloads/Telecomunication_data_source.xlsx'
Telecommunication_Industry = pd.read_excel(file_path)
```

```python
Telecommunication_Industry.head()
```

```python
#checking for missing values
for col in Telecommunication_Industry.columns:
    missing_values = np.mean(Telecommunication_Industry[col].isnull())
    print('{} - {}%'.format(col, missing_values))
```

```python
# Replacing missing values with a zero is a better choice for these columns
Telecommunication_Industry['Start'].fillna(0, inplace=True)
Telecommunication_Industry['End'].fillna(0, inplace=True)
Telecommunication_Industry['Total UL (Bytes)'].fillna(0, inplace=True)
Telecommunication_Industry['Total DL (Bytes)'].fillna(0, inplace=True)
Telecommunication_Industry['Dur. (ms)'].fillna(0, inplace=True)
```

```python
#verifying missing values conversion
for col in Telecommunication_Industry.columns:
    missing_values = np.mean(Telecommunication_Industry[col].isnull())
    print('{} - {}%'.format(col, missing_values))
    
```

## Calculating duration of the session for each user


```python
Telecommunication_Industry.groupby('MSISDN/Number')['Dur. (ms)'].sum()
```

## Calculating total upload and download for each user

```python
DL_and_UL_result = Telecommunication_Industry.groupby('MSISDN/Number').agg({'Total UL (Bytes)': 'sum', 'Total DL (Bytes)': 'sum'})
DL_and_UL_result
```

## Calculating session frequency for each user


```python
Dur_grouped = Telecommunication_Industry.groupby('MSISDN/Number').count()
Dur_grouped
#[Dur_grouped['Start']>5]
```

## Calculating top 10 users

```python
Telecommunication_Industry['MSISDN/Number'].value_counts().head(10)
```

```python
# Creating a new column adding Total DL and Total UL
Telecommunication_Industry['Total_Data_Volume'] = Telecommunication_Industry[['Total UL (Bytes)', 'Total DL (Bytes)']].sum(axis=1)
Telecommunication_Industry.head()
```

## Adding download and upload of applications 

```python
google = Telecommunication_Industry['Google DL (Bytes)']+ Telecommunication_Industry['Google UL (Bytes)']
email = Telecommunication_Industry['Email DL (Bytes)']+ Telecommunication_Industry['Email UL (Bytes)']
gaming = Telecommunication_Industry['Gaming DL (Bytes)']+ Telecommunication_Industry['Gaming UL (Bytes)']
youtube = Telecommunication_Industry['Youtube DL (Bytes)']+ Telecommunication_Industry['Youtube UL (Bytes)']
netflix = Telecommunication_Industry['Netflix DL (Bytes)']+ Telecommunication_Industry['Netflix UL (Bytes)']
social = Telecommunication_Industry['Social Media DL (Bytes)']+ Telecommunication_Industry['Social Media UL (Bytes)']

Telecommunication_Industry['google']=google
Telecommunication_Industry['email']=email
Telecommunication_Industry['gaming']=gaming
Telecommunication_Industry['youtube']=youtube
Telecommunication_Industry['netflix']=netflix
Telecommunication_Industry['social']=social
```
